## Load dataset

```python
# Read the raw data from TLC

yellow_taxi = pd.read_csv("https://s3.amazonaws.com/nyc-tlc/trip+data/yellow_tripdata_2016-06.csv")
```

```python
# Collect all the points roughly inside the bounding box of Manhattan (40.7, -74.02) and (40.85,-73.93)

yellow_taxi = yellow_taxi[((yellow_taxi.pickup_longitude >= -74.02) & (yellow_taxi.pickup_latitude >= 40.7) & \
                           (yellow_taxi.pickup_longitude <= -73.93) & (yellow_taxi.pickup_latitude <= 40.85))]
```

```python
# Calculate the trip duration and vehicle speed

def convert_to_unix(s):
    return time.mktime(datetime.datetime.strptime(s, "%Y-%m-%d %H:%M:%S").timetuple())

def select_data(yellow_taxi):
    duration = yellow_taxi[['tpep_pickup_datetime', 'tpep_dropoff_datetime']]
    # Pickups and dropoffs to unix time
    duration_pickup = [convert_to_unix(x) for x in duration['tpep_pickup_datetime'].values]
    duration_dropoff = [convert_to_unix(x) for x in duration['tpep_dropoff_datetime'].values]
    # Calculate duration of trips in minutes
    durations = (np.array(duration_dropoff) - np.array(duration_pickup)) / 60.0
    
    # Return a new data frame which contains the columns I am interested
    df = yellow_taxi[['passenger_count', 'trip_distance', 'pickup_longitude', 'pickup_latitude', 'dropoff_longitude', \
                      'dropoff_latitude', 'total_amount', 'tpep_pickup_datetime']]
    
    df['trip_times'] = durations
    df['pickup_times'] = duration_pickup  # Used for time binning later
    df['speed'] = 60.0 * (df['trip_distance'] / df['trip_times'])
    
    return df

yellow_taxi = select_data(yellow_taxi)
```

```python
# Extract the pickup day in the first full week (Mon-Sun) of June 2016 (June 13-19, 2016)

yellow_taxi['tpep_pickup_datetime']= pd.to_datetime(yellow_taxi['tpep_pickup_datetime'])
yellow_taxi['pickup_day'] = yellow_taxi['tpep_pickup_datetime'].dt.day
yellow_taxi['pickup_hour'] = yellow_taxi['tpep_pickup_datetime'].dt.hour
yellow_taxi_week = yellow_taxi[(yellow_taxi['pickup_day'] >= 13) & (yellow_taxi['pickup_day'] <= 19)]
```

## Data analysis

```python
# Pick-up location clustering

pickup_loc = yellow_taxi_week_modified[['pickup_latitude', 'pickup_longitude']].values
```

```python
# Cross-validation on K-means

def find_min_dist(cluster_center, cluster_len):
    min_dist = float('inf')
    
    for i in range(0, cluster_len):
        for j in range(0, cluster_len):
            if j != i:
                dist = mpu.haversine_distance((cluster_center[i][0], cluster_center[i][1]), (cluster_center[j][0], cluster_center[j][1]))
                min_dist = min(min_dist, dist/1.60934)
        
    print("Cluster size of: ", cluster_len)
    print("Inter-cluster distance minimum: ", min_dist)
    print()
```

```python
# Implement K-means to cluster Manhattan

def find_cluster(num):
    kmeans = KMeans(n_clusters = num, random_state = 123)
    kmeans.fit(pickup_loc)
    cluster_center = kmeans.cluster_centers_
    cluster_len = len(cluster_center)
    return cluster_center, cluster_len
```

```python
# Cross-validation for [10, 15, 20, 25, 30, 35]

for num in range(10, 40, 5):
    cluster_center, cluster_len = find_cluster(num)
    find_min_dist(cluster_center, cluster_len)
```

```python
# Select 15 clusters to solve the problem

kmeans = KMeans(n_clusters = 15, random_state = 123)
kmeans.fit(pickup_loc)
yellow_taxi_week_modified['pickup_cluster'] = kmeans.predict(yellow_taxi_week_modified[['pickup_latitude', 'pickup_longitude']])
cluster_center = kmeans.cluster_centers_
cluster_len = len(cluster_center)
```

```python
# Plot the cluster

def plot_cluster(data):
    city_long_border = (-74.03, -73.93)
    city_lat_border = (40.69, 40.84)
    fig, ax = plt.subplots(ncols = 1, nrows = 1, figsize = (10, 15))
    ax.scatter(data.pickup_longitude.values, data.pickup_latitude.values, s = 15, lw = 0, c = data.pickup_cluster.values, cmap = 'tab20', alpha = 0.8)
    ax.set_xlim(city_long_border)
    ax.set_ylim(city_lat_border)
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    plt.show()
    
plot_cluster(yellow_taxi_week_modified)
```

```python
# Location-wise average efficiency over the time period

cluster_num = set(yellow_taxi_week_modified['pickup_cluster'])
cluster_eff = {}
location_eff = []
num_per_cluster = {}
num_within = []
cluster = np.array(yellow_taxi_week_modified['pickup_cluster'])

for i in cluster_num:
    yellow_taxi_cluster = yellow_taxi_week_modified[(yellow_taxi_week_modified.pickup_cluster == i)]
    cluster_eff[i] = yellow_taxi_cluster['charge_per_mile_per_passenger'].mean()
    num_per_cluster[i] = len(yellow_taxi_cluster)

for i in range(len(cluster)):
    location_eff.append(cluster_eff[cluster[i]])
    num_within.append(num_per_cluster[cluster[i]])
    
yellow_taxi_week_modified['cluster_eff'] = np.array(location_eff)
yellow_taxi_week_modified['num_per_cluster'] = np.array(num_within)

# Only consider the clusters with over 50 data points

yellow_taxi_week_modified = yellow_taxi_week_modified[(yellow_taxi_week_modified.num_per_cluster >= 50)]
```

```python
# Time binning

unix_pickup_times = [i for i in yellow_taxi_week_modified['pickup_times'].values]
daywise_binned_unix_pickup_times = [int((i-1465790400)/86400)+1 for i in unix_pickup_times]
print("Min = {}, Max = {}".format(min(daywise_binned_unix_pickup_times), max(daywise_binned_unix_pickup_times)))

hourwise_binned_unix_pickup_times = [int((i-1465790400)/3600)+1 for i in unix_pickup_times]
print("Min = {}, Max = {}".format(min(hourwise_binned_unix_pickup_times), max(hourwise_binned_unix_pickup_times)))
```

```python
yellow_taxi_week_modified['day_pickup_bins'] = np.array(daywise_binned_unix_pickup_times)
yellow_taxi_week_modified['hour_pickup_bins'] = np.array(hourwise_binned_unix_pickup_times)
```

```python
del unix_pickup_times
del daywise_binned_unix_pickup_times
del hourwise_binned_unix_pickup_times
```

```python
# Average economic efficiency for each time bin

yellow_taxi_week_modified_daybin = yellow_taxi_week_modified.groupby(['pickup_cluster', 'day_pickup_bins']).charge_per_mile_per_passenger.agg({'day_eff': 'mean'})
yellow_taxi_week_modified_hourbin = yellow_taxi_week_modified.groupby(['pickup_cluster', 'hour_pickup_bins']).charge_per_mile_per_passenger.agg({'hour_eff': 'mean'})
```

```python
# Plot the location-wise efficiency

def plot_cluster_eff(data):
    city_long_border = (-74.03, -73.93)
    city_lat_border = (40.69, 40.84)
    plt.figure(figsize = (10, 15))
    scale_map = plt.get_cmap('YlOrRd')
    cluster = plt.scatter(data.pickup_longitude.values, data.pickup_latitude.values, s = 15, lw = 0, c = data.cluster_eff.values, cmap = scale_map)
    plt.xlim(city_long_border)
    plt.ylim(city_lat_border)
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.colorbar(cluster)
    # plt.savefig('Mon.png', dpi = 200)
    plt.show()
    
plot_cluster_eff(yellow_taxi_week_modified)
```

```python
# Daily efficiency

day_eff = {}

for i in range(7):
    yellow_taxi_cluster = yellow_taxi_week_modified[(yellow_taxi_week_modified.pickup_day == i + 13)]
    day_eff[i + 13] = yellow_taxi_cluster['charge_per_mile_per_passenger'].mean()
```

```python
# Plot daily efficiency

x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
plt.bar(x, day_eff.values(), color = 'turquoise', width = 0.55)
plt.ylabel("Average efficiency in Manhattan")
# plt.savefig('Figure-2.eps')
```
